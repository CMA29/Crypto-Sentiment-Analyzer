Use the news api to pull the latest news articles for bitcoin and ethereum and create a DataFrame of sentiment scores for each coin. 

Use descriptive statistics to answer the following questions:

```python
# Initial imports
import os
import json
import pandas as pd
from dotenv import load_dotenv
import nltk
from nltk.sentiment.vader import SentimentIntensityAnalyzer
analyzer = SentimentIntensityAnalyzer()
load_dotenv()

%matplotlib inline
```

```python
# Fetch the Bitcoin news articles
bit_news = newsapi.get_everything(q="bitcoin",language="en")


with open('bit.json', 'w') as json_file:
    json.dump(bit_news, json_file)
    
```

```python
# Fetch the Ethereum news articles
eth_news = newsapi.get_everything(q='ethereum',language='en')


with open('eth.json', 'w') as json_file:
    json.dump(eth_news, json_file)

```

```python
# Create the Bitcoin sentiment scores DataFrame

bit_df = pd.DataFrame.from_dict(bit_news["articles"])
bit_df = bit_df.drop(['author', 'title', 'url','urlToImage','publishedAt','content',], axis=1)
sentiment_title = bit_df.apply(lambda x : analyzer.polarity_scores(x['description']), axis=1)
bit_df['Comp'] =sentiment_title.apply(lambda x : x.get('compound') )
bit_df['Pos'] =sentiment_title.apply(lambda x : x.get('pos') )
bit_df['Neu'] =sentiment_title.apply(lambda x : x.get('neu') )
bit_df['Neg'] =sentiment_title.apply(lambda x : x.get('neg') )

bit_df.head()
```

```python
# Create the ethereum sentiment scores DataFrame
eth_df = pd.DataFrame.from_dict(eth_news["articles"])
eth_df = eth_df.drop(['author', 'title', 'url','urlToImage','publishedAt','content',], axis=1)
sentiment_title = eth_df.apply(lambda x : analyzer.polarity_scores(x['description']), axis=1)
eth_df['Comp'] =sentiment_title.apply(lambda x : x.get('compound') )
eth_df['Pos'] =sentiment_title.apply(lambda x : x.get('pos') )
eth_df['Neu'] =sentiment_title.apply(lambda x : x.get('neu') )
eth_df['Neg'] =sentiment_title.apply(lambda x : x.get('neg') )



eth_df.head()
```

# Tokenizer

In this section, you will use NLTK and Python to tokenize the text for each coin. Be sure to:
1. Lowercase each word
2. Remove Punctuation
3. Remove Stopwords

```python
from nltk.tokenize import word_tokenize, sent_tokenize
from nltk.corpus import stopwords
from nltk.stem import WordNetLemmatizer, PorterStemmer
from string import punctuation
import re
```

```python
# Expand the default stopwords list if necessary
nltk.download('wordnet')
```

```python
# Complete the tokenizer function
def tokenizer(text):
    """Tokenizes text."""

    lemmatizer = WordNetLemmatizer()
    sw = set(stopwords.words('english'))
    regex = re.compile("[^a-zA-Z ]")
    re_clean = regex.sub('', text)
    words = word_tokenize(re_clean)
    return [lemmatizer.lemmatize(word.lower()) for word in words if word.lower() not in set(stopwords.words('english'))]
    
```

```python
# Create a new tokens column for bitcoin
import string
bit_df['tokens'] = bit_df.apply(lambda x : tokenizer(x['description']), axis=1)
bit_df.head()
```

```python
# Create a new tokens column for ethereum
import string
eth_df['tokens'] = eth_df.apply(lambda x : tokenizer(x['description']), axis=1)
eth_df.head()
```

# NGrams and Frequency Analysis

In this section you will look at the ngrams and word frequency for each coin. 

1. Use NLTK to produce the n-grams for N = 2. 
2. List the top 10 words for each coin. 

```python
from collections import Counter
from nltk import ngrams
```

```python
# Generate the Bitcoin N-grams where N=2
flat_bit_tokens = [item for sublist in bit_df.tokens.to_list() for item in sublist]
bigram_counts = Counter(ngrams(flat_bit_tokens, n=2))
bigram_counts.most_common(20)
```

```python
# Generate the Ethereum N-grams where N=2
flat_eth_tokens = [item for sublist in eth_df.tokens.to_list() for item in sublist]
eth_bigram_counts = Counter(ngrams(flat_eth_tokens, n=2))
eth_bigram_counts.most_common(20)
```

```python
# Use the token_count function to generate the top 10 words from each coin
def token_count(tokens, N=10):
    """Returns the top N tokens from the frequency count"""
    return Counter(tokens).most_common(N)
```

```python
# Get the top 10 words for Bitcoin
token_count(flat_bit_tokens)
```

```python
# Get the top 10 words for Ethereum
token_count(flat_eth_tokens)
```
